#PRÉ-PROCESSAMENTO

---



```python
generos = []

for row in range(0, len(df)):
  generos.append(str(df['genero'].iloc[row]).split(u',\xa0'))

df['genero'] = generos
```

```python
df['genero'].head()
```

```python
generos = []

for i in range(0, len(df)):
  for j in range(0, len(df['genero'].iloc[i])):
    if df['genero'].iloc[i][j] not in generos:
      generos.append(df['genero'].iloc[i][j])

print(generos) #reutilizando variavel para listar todos os gêneros
```

```python
filme_genero = {}

for key in generos:
    filmes = []
    for value in range(0, len(df)):
        if key in df['genero'].iloc[value]:
          filmes.append(value)

    filme_genero[key] = filmes

print(filme_genero)
```

```python
print(
    df.iloc[filme_genero['Animation'][0]]
)
```

##gêneros tratados, normalizando notas em 3 categorias

```python
desempenho_nota = []

for nota in df['score']:
  if nota <= 33.3:
    desempenho_nota.append('Baixa')
  elif nota <= 66.6:
    desempenho_nota.append('Média')
  else:
    desempenho_nota.append('Alta')

df['desempenho'] = desempenho_nota
```

```python
generos_media = {}

for genero in generos: #Forzinho bacana
  generos_media[genero] = df[df['genero'].apply(lambda x: genero in x)]['score'].mean()


df_generos_media = pd.DataFrame(list(generos_media.items()), columns = ['genero', 'media']) #E assim cria-se um dataframe
```

```python
df_generos_media
```

```python
ax = sns.barplot(x='media', y='genero', data = df_generos_media, palette = Figura)
plt.rcParams['figure.figsize'] = (10, 10) #Aumentar o tamanho da Figura
plt.rcParams.update({"font.size": 18}) #Aumentar a letra
plt.axvline(x=df_generos_media['media'].mean(), color='red', linestyle='--', linewidth=2, label='Média') #Média linha vermelha
plt.savefig('Linha_x_avaliacao.png', dpi=300)
```

```python
orcamento_medio = {}

for genero in generos:
  orcamento_medio[genero] = df[df['genero'].apply(lambda x: genero in x)]['orçamento'].mean()


df_orcamento_medio = pd.DataFrame(list(orcamento_medio.items()), columns = ['genero', 'orçamento_medio'])
```

```python
ax = sns.barplot(x='orçamento_medio', y='genero', data = df_orcamento_medio, palette = Figura)
plt.rcParams['figure.figsize'] = (10, 10) #Aumentar o tamanho da Figura
plt.rcParams.update({"font.size": 18}) #Aumentar a letra
plt.axvline(x=df_orcamento_medio['orçamento_medio'].mean(), color='red', linestyle='--', linewidth=2, label='Média') #Média linha vermelha
plt.savefig('Linha_x_avaliacao.png', dpi=300)
```

```python
bilheteria_media = {}

for genero in generos:
  bilheteria_media[genero] = df[df['genero'].apply(lambda x: genero in x)]['bilheteria'].mean()


df_bilheteria_media = pd.DataFrame(list(bilheteria_media.items()), columns = ['genero', 'bilheteria_media'])
```

```python
ax = sns.barplot(x='bilheteria_media', y='genero', data = df_bilheteria_media, palette = Figura)
plt.rcParams['figure.figsize'] = (10, 10) #Aumentar o tamanho da Figura
plt.rcParams.update({"font.size": 18}) #Aumentar a letra
plt.axvline(x=df_bilheteria_media['bilheteria_media'].mean(), color='red', linestyle='--', linewidth=2, label='Média') #Média linha vermelha
plt.savefig('Linha_x_avaliacao.png', dpi=300)
```
